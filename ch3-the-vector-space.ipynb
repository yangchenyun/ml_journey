## Definitions

### Definition 3.1.1 Linear Combination
Suppose $v1, . . . , vn$ are vectors. We define a __linear combination__ of v1, . . . , vn to be a sum of
$\alpha_1*v_1 + \alpha_2*v_2 + ... + \alpha_n*v_n$, where $\alpha_1, ... \alpha_n$ are scalars and referred to as 
__coefficients__.

If all coefficients are zero, it is a __trivial linear combination__.

A linear system implies other equations.

### Definition 3.2.1 Span
The set of *all* linear combinations of vectors $v1 , . . . , vn$ is called the span of these vectors, and is written $Span \{v1 , . . . , vn \}$.

### Definition3.2.9 Generator
Let $V$ be a set of vectors (also denoted as $F^D$). If $v1,...,vn$ are vectors such that $V=Span\{v1,...,vn\}$ then we say $\{v1,...,vn\}$ is a generating set for V, and we refer to the vectors v1,...,vn as generators for V.

For any finite domain D, there are __standard generators__ $E$ for $F^D$, which could be defined as $\forall k \in D, E \space is \space e_k = \{k: 1\}$

### Definition flat
A flat is an affine space in R^n for some n.
It is any non-empty subset of W that is closed under vector addition and scalar multiplication. 

A flat is a term used to refer to a subspace of a higher-dimensional space that has lower dimension. 

### Definition 3.4.1 Vector space
A set V of vectors is called a vector space if it satisfies Properties V1, V2, and V3
- V1: V contains zero vector
- V2: V is closed under scalar-vector multiplication
- V3: V is closed under vector addition

F^D is a vector space too.

Both span of vectors and homogeneous linear system defines vector spaces.

#### Definition 3.4.7 trivial vector space
A vector space consisting only of a zero vector is a __trivial vector space__.

#### Definition 3.4.9 subspace
If V and W are vector spaces and V is a subset of W, we say V is a subspace of W.

$\mathbb{R}^2$ subspaces include:
- $\{0^2\}$, zero vector
- $\mathbb{R}^2$
- For any nonzero vector $\{a, b\}$, $span \{[a, b]\}$, (i.e. a line defines one vector space)

$\forall span \{[a, b], [c, d]\}, \text{where } \{[c, d]\} \notin span \{[a, b]\}, span \{[a, b], [c, d]\} \equiv \mathbb{R}^2$
Two basis lines define the vector space of $\mathbb{R}^2$.

### Definition 3.5.2 affine combination
A linear combination $α_1u_1 + · · · + α_nu_n$ is called an affine combination if the coefficients sum to one.

Set of __all__ affine combination of a collections of vectors is __affine hull__ of that collections.

The affine hull of $u_1, u_2, ...$ is equivalent to $span \{u_2-u_1, u_3 - u_1 ... u_n - u_1\}$


### Definition 3.5.8 affine space
An affine space is the result of translating a vector space. That is, a set A is an affine space if there is a vector a and a vector space V such that:

$A = \{ a + v : v \in V \}$

A affine space could also be represented as the affine hull of vectors.


### Theorem 3.6.2: 
For any linear system, the set of solutions either is empty or is an affine space.

The solution to original linear system is an affine space; every affine space is a soluiton set of a linear system.

The number of solution equals to the cardinality of the vector space consists of solutions to the corresponding homogeneous system.

#### Corollary 3.6.4
Suppose a linear system has a solution. The solution is unique if and only if the only solution to the corresponding homogeneous linear system is the zero vector.

## Representation of flats
A point in n-vector over $\mathbb{R}$ space could be represented as:
- The affine space for 0 n-vector. Intuition: the translation of a zero vector is a point.
- A linear system soluiton with n equations: $\alpha_1 . x = \beta_1, \alpha_2 . x = \beta_2, ... \alpha_n . x = \beta_n$

A line in n-vector over $\mathbb{R}$ space could be represented as:
- The affine space for 1 n-vector. Intuition, any line could be defined as the translation of a line through origin (vector-space).
- A linear system soluiton with n equations: $\alpha_1 . x = \beta_1, \alpha_2 . x = \beta_2, ... \alpha_n . x = \beta_{n-1}$

A plane in n-vector over $\mathbb{R}$ space could be represented as:
- The affine space for 2 n-vector. Intuition, any plane could be defined as the translation of a plane through origin (vector-space).
- A linear system soluiton with n equations: $\alpha_1 . x = \beta_1, \alpha_2 . x = \beta_2, ... \alpha_n . x = \beta_{n-2}$

## 3.3 The geometry of sets of vectors

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_span_3d(v1, v2, n_samples=10, marker='x', color='r', size=20):
    # Calculate the span of v1 and v2
    span = np.array([v1, v2]).T
    basis = np.linalg.qr(span)[0]

    alpha_beta = np.linspace(-5, 5, num=n_samples)
    alpha_beta = np.transpose([np.tile(alpha_beta, len(alpha_beta)), np.repeat(alpha_beta, len(alpha_beta))])


    # Calculate the linear combinations of v1 and v2 with the random coefficients
    points = np.dot(basis, alpha_beta.T).T

    # Plot the points in 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(points[:,0], points[:,1], points[:,2], marker=marker, c=color, s=size)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()

v1 = np.array([1, 0, 1.65])
v2 = np.array([0, 1, 1])
plot_span_3d(v1, v2)
```

```python
# Definition 3.3.8: A linear equation with right-hand side zero is a homogeneous linear equation.
# Definition 3.3.11: A linear system (collection of linear equations) with all right-hand sides zero is called 
# a homogeneous linear system. (homogeneous, all the right-hands are all zero.)

# Hypothesis 3.3.12: A flat containing the origin is the solution set of a homogeneous linear system.

# Use the span of a vector to define a line
def plot_lines(vectors, n_samples=20, sizes=None):
    if sizes is None:
        sizes = [20] * len(vectors)

    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    for i, v in enumerate(vectors):
        # Generate a set of coefficients for v
        alpha = np.linspace(-10, 10, num=n_samples)

        # Calculate the linear combinations of v with the coefficients
        points = alpha[:,np.newaxis] * v[np.newaxis,:]

        # Plot the points in 3D
        color = colors[i]
        ax.plot(points[:,0], points[:,1], points[:,2], c=color)
        ax.scatter(points[:,0], points[:,1], points[:,2], c=color, s=sizes[i])

    # Add labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()

vectors = [np.array([4, -1, 1]), np.array([0, 1, 1])]
sizes = [20, 20]
plot_lines(vectors, n_samples=20, sizes=sizes)
```

```python
def plot_plane(v, n_samples=20):
    # Find two orthogonal vectors that span the plane
    v2, v3, _ = np.eye(3)
    v2 -= v * np.dot(v, v2)
    v2 /= np.linalg.norm(v2)
    v3 -= v * np.dot(v, v3)
    v3 -= v2 * np.dot(v2, v3)
    v3 /= np.linalg.norm(v3)

    # Generate a set of points on the plane
    x, y = np.meshgrid(np.linspace(-1, 1, n_samples), np.linspace(-1, 1, n_samples))
    z = -(v[0]*x + v[1]*y) / v[2]

    # Plot the plane
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.5, color='b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Plane defined by vector {}'.format(v))
    plt.show()

# Example usage:
v = np.array([4, -1, 1])
plot_plane(v, n_samples=20)
plot_plane(np.array([0, 1, 1]), n_samples=20)
```

```python
# Application - checksum
# A function could detect a random change
import random
from book.vec import Vec
from book.GF2 import one, zero

def to_vector(l):
    """Accepts binary string or list for GF(2) vector."""
    f = dict(enumerate(l))
    return Vec(
        set(f.keys()), # D is [0, 1, 2, ...]
        f)

def n_vector(n):
    return to_vector([one if random.uniform(0, 1) > 0.5 else 0 for _ in range(n)])

def gf2_hex(v):
    """converts vector in GF(2) field to hex string."""
    gf_to_str = lambda v: '1' if v == one else '0'
    binary_string = ''.join(map(gf_to_str, v))  # Convert binary list to string
    decimal_num = int(binary_string, 2)  # Convert binary string to decimal number
    hex_string = hex(decimal_num)[2:]  # Convert decimal number to hexadecimal string (and remove '0x' prefix)
    return hex_string

# file is a n-vector over GF(2)
n = 64
file_content = n_vector(n)

# vector used in checksum
# TODO: (isn't it large?)
check_vectors = [n_vector(n) for _ in range(n)]

def checksum(file, vectors):
    return [v * file for v in vectors]

gf2_hex(checksum(file_content, check_vectors))
```

```python
import functools

# Problem 3.8.1
def vec_random(D, generator):
    """Generate a vector D using a random generator in field."""
    f = {d: generator(d) for d in D}
    return Vec(D, f)

vec_gauss = lambda D: vec_random(D, lambda x: int(random.gauss(0, 4)))

def vec_select(veclist, k):
    return [v for v in veclist if v[k] == 0]

def vec_sum(veclist, D):
    return functools.reduce(lambda acc, v: acc + v, veclist, Vec(D, {}))

def vec_select_sum(D, veclist, k):
    return vec_sum(vec_select(veclist, k), D)

D = set(range(10))
veclist = [vec_gauss(D) for _ in range(10)]

[
    "vec_select",
    vec_select(veclist, 0),
    vec_select(veclist, 1),
    vec_select(veclist, 2),
    "vec_sum",
    vec_sum(veclist, D),
    f"vec_select_sum({D}, veclist, 6)",
    vec_select(veclist, 6),
    vec_select_sum(D, veclist, 6),
]
```

```python
import itertools
# Problem 3.8.10
# My intuition is that even number seems to be vector space; 
# Odd number isn't
# - zero 5-vectors is not in the space; 
# - vector addition is not closed "11100" + "10000" => "01100"

def is_even_ones(v):
    return len([v[d] for d in v.D if v[d] == one]) % 2 == 0

def is_odd_ones(v):
    return not is_even_ones(v)

vec_gf = lambda D: vec_random(D, lambda x: one if random.uniform(0, 1) > 0.5 else zero)

def g():
    while True:
        yield vec_gf(set(range(5)))

# Experiment validates the results
odd_space = list(itertools.islice(itertools.filterfalse(is_even_ones, g()), 25))
even_space = list(itertools.islice(itertools.filterfalse(is_odd_ones, g()), 25))

[
    "is odd_space closed on vector add",
    all(is_odd_ones(u) for u in
        [v[0] + v[1] for v in itertools.product(odd_space, repeat=2)]),
    "is odd_space closed on vector scalar mul",
    all(is_odd_ones(u) for u in
        [v[0]*v[1] for v in itertools.product([one, zero], odd_space)]),
    "is even_space closed on vector add",
    all(is_even_ones(u) for u in
        [v[0] + v[1] for v in itertools.product(even_space, repeat=2)]),
    "is even_space closed on vector scalar mul",
    all(is_even_ones(u) for u in
        [v[0]*v[1] for v in itertools.product([one, zero], even_space)]),
]
```
