# Compare forward, backward, and central differences for numerically computing  derivatives

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
dt = .2
t = np.linspace(-2, 4, 60)
f = np.sin(t)
```

```python
# Exact Derivative
dfdt = np.cos(t)
```

```python
# plotting commands
plt.plot(t, f, 'k--', linewidth=1.2)
plt.grid(True)
plt.plot(t, dfdt, 'k', linewidth=3)
plt.legend(['Function', 'Exact Derivative'], fontsize=14)
plt.xlim(-2, 4)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
# Forward Difference Approximation
dfdtF = (np.sin(t + dt) - np.sin(t)) / dt

# Backward Difference Approximation
dfdtB = (np.sin(t) - np.sin(t - dt)) / dt

# Central Difference Approximation
dfdtC = (np.sin(t + dt) - np.sin(t - dt)) / (2 * dt)
```

```python
plt.figure(figsize=(20, 4))
plt.subplot(1, 2, 1)
plt.plot(t, dfdt, 'k', linewidth=3)
plt.plot(t, dfdtF, 'b', linewidth=1.2)  # Forward Difference
plt.plot(t, dfdtB, 'g', linewidth=1.2)  # Backward Difference
plt.plot(t, dfdtC, 'r', linewidth=1.2)  # Central Difference
plt.legend(['Exact Derivative', 'Forward Diff', 'Backward Diff', 'Central Diff'], 
       fontsize=14, framealpha=1.0)  # get rid of legend transparency
plt.grid(True)

# now make a zoomed-in plot
plt.subplot(1, 2, 2)
plt.plot(t, dfdt, 'k', linewidth=3)
plt.plot(t, dfdtF, 'b', linewidth=1.2)  # Forward Difference
plt.plot(t, dfdtB, 'g', linewidth=1.2)  # Backward Difference
plt.plot(t, dfdtC, 'r', linewidth=1.2)  # Central Difference
plt.grid(True)
plt.xlim(0, 2)
plt.show()
```
