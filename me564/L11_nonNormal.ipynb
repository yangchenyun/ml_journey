# Investigation of nonnormal linear growth

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
```

```python
t = np.linspace(0, 1000, 10000)
A = np.array([[-0.009, 1], [0, -0.01]])
y0 = np.array([0, 1])

def linear_ode(t, y):
    return A @ y

# integrate linear system from (0, 1000) with 
# initial condition of y0 and returning solution at times t_eval
linear_ode_solution = solve_ivp(linear_ode, (0, 1000), y0, t_eval=t)
y = linear_ode_solution.y.T
plt.plot(t, y)
plt.legend(['x', 'v'], fontsize=12, framealpha=1.0)
plt.ylabel('x, v')
plt.xlabel('Time')
plt.grid(True)
```

```python
# Repeat analysis for slightly different system matrix A
t = np.linspace(0, 20, 2000)
A = np.array([[-1, 1], [0, -1]])
linear_ode_solution = solve_ivp(linear_ode, (0, 20), y0, t_eval=t)
y = linear_ode_solution.y.T
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('x,v')
plt.legend(['x', 'v'], fontsize=12, framealpha=1.0)
plt.grid(True)
plt.show()
```
