# Chapter 4 - The Matrix 

#### Notation
Adapt pythonic notation:
$M[r, c], M[:, c], M[r, :]$ or $M_{r,c}, M_{:,c}, M_{r,:}$

#### Definition 4.1.6 Identity Matrix
For a finite set D, the D × D identity matrix is the matrix whose row-label set and column-label set are both D, and in which entry (d, d) is a 1 for every d ∈ D (and all other entries are zero).

Row-addition matrix, a matrix that is an identity matrix plus at most one off-diagonal nonzero entry.

#### Definition 4.2.1 Column space and row space
For a matrix M
• the column space of M, written Col M, is the vector space spanned by the columns of M, and
• row space of M, written Row M, is the vector space spanned by the rows of M.

#### Definition 4.2.1 Transpose
A matrix M is a symmetric matrix if $M^T = M$.

## Operation between Matrix and Vector

These operations could be interpreted as matrix operations between matrix and column-vector.

#### Definition 4.5.6 vector-matrix, matrix-vector multiplication
$M*v = \sum_{c \in C} v[c] * M[:,c]$, vector is used as coefficients for the column vectors in matrix.

$w*M = \sum_{r \in R} w[r] * M[r,:]$, vector is used as coefficients for the row vectors in matrix.


$w*M \equiv M^T*w$


The multiplication between vector and matrix is not communitative.

#### Definition 4.6.1 (Dot-Product Definition of Matrix-Vector Multiplication)

If M is an R×C matrix and u is a C-vector then M ∗u is the R-vector v such that v[r] is the dot-product of __row r of M__ with u.

#### Definition 4.6.3 (Dot-Product Definition of Vector-Matrix Multiplication)
If M is an R×C matrix and u is a R-vector then u∗M is the C-vector v such that v[c] is the dot-product of u __with column c of M__.Then the system of linear equations is equivalent to the matrix-vector equation A ∗ x = b.

#### Implementation for sparse

#### Linear System Representation
$A * x = b$

Solving a linear system becomes solving a matrix equation.

#### Definition 4.6.9 upper - triangular matrix
An n × n upper-triangular matrix A is a matrix with the property that $A_{ij} =0, i > j$

To generalize to our matrices with arbitrary row- and column-label sets, we specify orderings of the label-sets. $A[LR[i], LC [j]] = 0, i > j$

#### Definition 4.10.20 Diagonal matrix
For a domain D, a D × D matrix M is a diagonal matrix if M [r, c] = 0 for every pair r, c ∈ D such that r ̸= c

#### Proposition 4.6.13
- $M * (\alpha \mathbf{v}) = \alpha(M * \mathbf{v})$
- $M * (\mathbf{u} + \mathbf{v}) = M * \mathbf{u} + M * \mathbf{v} $

#### Definition 4.7.1 Null space
The null space of a matrix A is the set ${v : A ∗ v = 0}$. It is written Null A.

Null space is a vector space.

Lemma 4.7.4

For any R × C matrix A and C-vector v, a vector z is in the null space of A if and only if A ∗ (v + z) = A ∗ v.

Corollary 4.7.6

Suppose a matrix-vector equation Ax = b has a solution. The solution is unique if and only if the null space of A consists solely of the zero vector.

## Matrix and Functions
Given M is R x C matric over F: $F_M: \mathbb{F}^C -> \mathbb{F}^R, f_M(\mathbf{x}) = M * \mathbf{x}$

In reverse, given $F_M: \mathbb{F}^A -> \mathbb{F}^B$.
M must be B x A, each column equals $f_M(e_a)$, $e_a$ is the standard generator for A.

The intuition here is use the basic vector to determine the matrix value. 

### Lemma 4.10.19

If $f : F^C → F^R$ is a linear function then there is an R × C matrix M over F such that $f(x) = M*x, \forall x \in F^C$.

### Linear functions
Definition by two properties:
- $f(\alpha u) = \alpha f(u)$
- $f(u + v) = f(u) + f(v)$

#### Lemma 4.10.10: 
If f : U −→ V is a linear function then f maps the zero vector of U to the zero vector of V.

#### Definition 4.10.11 kernel of linear function
The kernel of a linear function f to be {v : f(v) = 0}. We denote the kernel of f by $Ker f$.

#### Lemma 4.10.12 
$Ker f$ is also a vector space.


#### Proposition 4.10.14
For a linear function f, for any vectors u1, . . . , un in the domain of f and any scalars α1,...,αn,

$f(α1u1 +···+αnun)=α1f(u1)+···+αnf(un)$

#### Lemma 4.10.15 (One-to-One Lemma)
A linear function is one-to-one __if and only if__ its kernel is a trivial vector space.

#### Lemma 4.10.18
The image of a linear function is __a subspace__ of the function’s co-domain

The key proof here is to prove it is a subspace.

## Matrix-matrix multiplication
Definition 4.11.1 Vector-matrix definition of matrix-matrix multiplication

$AB[r] = A[r] * b, \forall r \in A.D[0]$

Definition 4.11.3 Matrix-vector definition of matrix-matrix multiplication

$AB[c] = A * B[c], \forall c \in B.D[1]$


Definition 4.11.7 (Dot-product definition of matrix-matrix multiplication)

Entry rc of AB is the dot-product of row r of A with column c of B

#### Lemma 4.11.8 (Matrix-Multiplication Lemma): 
fAB = fA ◦ fB

#### Corollary 4.11.11
Matrix-matrix multiplication is associative.

Intuition, this is a transformation of the coordinate system.

#### Proposition 4.11.14
For matrices A and B, $(AB)^T =B^T*A^T$

## Inverse Matrix
Matrix inverse, defined by inverse functions.

identity inverse

elementary matrix inverse

#### Lemma 4.13.11
If the R × C matrix A has an inverse A−1 then AA−1 is the R × R identity matrix.

#### Proposition 4.13.12 invertible matrix in unique solution
If A is invertible then, for any vector b with domain equal to the row-label set of A, the matrix-vector equation Ax = b has exactly one solution, namely A^−1b

#### Lemma 4.13.13
Suppose A is an upper-triangular matrix. Then A is invertible if and only if none of its diagonal elements is zero.

#### Proposition 4.13.14
If A and B are invertible matrices and the matrix product AB is defined then AB is an invertible matrix, and (AB)^−1 = B^−1A^−1

Prove a matrix is inversible could be break down to find inversible factors.

#### Fallacy
It is tempting to conjecture that, conversely, for any matrix A, if B is a matrix such that AB is an identity matrix 1 then B is the inverse of A. This is not true.

#### Corollary 4.13.19
Matrices A and B are inverses of each other if and only if both AB and BA are identity matrices.

```python
# Example 4.6.4: Image Blurr
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('bot.webp')
new_size = (img.size[0] // 2, img.size[1] // 2)
img = img.resize(new_size)

# 256 * 256 * 3 array
img_arr = np.array(img)

# Display the image using Matplotlib
plt.imshow(img_arr)
plt.show()
```

```python
h = len(img_arr)
w = len(img_arr[0])
D_orig = {(r, c) for c in range(w) for r in range(h)}
scale = 4
D_low = {(r, c) for c in range(int(w/scale)) for r in range(int(h/scale))}

# In M, each row is a pixel in low resolution image where each column is a pixel in high resolution where
# only the relevant values have a value
M = Mat((D_low, D_orig),
        {((r, c), (4*r+m, 4*c+n)): 1.0/scale**2 for m in range(scale) for n in range(scale)
         for r in range(int(h/scale)) for c in range(int(w/scale))})
```

```python
u = Vec(D_orig, {(r, c): img_arr[r][c] for c in range(w) for r in range(h)})
```

```python
# This implementation takes forever
# O(n^2 * m), n = origin dimension, m = new dimension
v = M*u
```

```python
new_img = np.zeros((int(h/scale), int(w/scale), 3), dtype=int)
for k in v.f: new_img[k[0]][k[1]] = v[k]

# Display the image using Matplotlib
plt.imshow(new_img)
plt.show()
```

```python
# Example 4.6.6, kernel for audio
m = Matrix([0, 1, -1])

def wrap_around(M, n):
    """Convert a kernel into wrap-around matrix."""
    kernel = list(M)
    result = []
    for i in range(n):
        for j in range(n):
            kernel_index = (-i+j) % n
            if kernel_index < len(kernel):
                result.append(kernel[kernel_index])
            else:
                result.append(0)

    return Matrix(n, n, result)

wrap_around(m, 10)
```
